```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
print('positive correlation')
x=[4, 7, 12, 18, 25, 29]
y=[7,14, 19, 28, 29, 40]
plt.scatter(x, y, s=100)
plt.grid()
plt.plot([0, 29], [5, 40])
plt.show()
```

```python

#negative correlation
x=[4, 7, 12, 18, 25, 29]
y=[40, 29, 28, 19, 14, 7,]

plt.scatter(x, y, c='g', s=100)
plt.grid()
plt.plot([5, 35], [40, 0])
plt.show()
print('negative correlation')
```

```python
#no correlation
x=[23, 8, 1, 9, 0, 29]
y=[40, 29, 28, 1, 14, 21]

plt.scatter(x, y, c='g', s=100)
plt.grid()
plt.plot([5, 35], [40, 0])
plt.show()
print('no correlation')
```
